```python
from hat.archs import hat_arch
import yaml
import torch

# hat-s 4xsr weights
w = torch.load("./experiments/pretrained_models/HAT-S_SRx4.pth")

# hat-s 4xsr model options from yml
with open("./options/train/train_HAT-S_SRx4_finetune_from_SRx2.yml", 'r') as f:
    opts = yaml.safe_load(f)
nopts = opts["network_g"]
del nopts["type"]

model = hat_arch.HAT(**nopts)
model_ema = hat_arch.HAT(**nopts)
print(model.load_state_dict(w["params"], strict=False))
print(model_ema.load_state_dict(w["params_ema"], strict=False))

new_w = {
    "params": model.state_dict(),
    "params_ema": model_ema.state_dict()
}

torch.save(new_w, "./experiments/pretrained_models/HAT-S_SRx4_AST.pth")
```
